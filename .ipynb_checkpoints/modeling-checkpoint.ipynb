```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression
from sklearn import metrics

plt.style.use('seaborn')
df = pd.read_csv('data/housing_aspects.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
plt.figure(figsize=(15,12))
plt.title('Correlation')
#plt.colorbar().set_label('Strength', fontsize=14)
sns.heatmap(df.corr().abs(), center=0, annot=True, cmap='rocket', cbar_kws={'label': 'Strength'});

# Workaround for a bug that cut off half of the top and bottom cells
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show;
```

```python
df_log = df.copy(deep=True)
list(df_log.columns)
```

```python
#Log Transformations
logprice = np.log(df_log['price'])
logbedrooms = np.log(df_log['bedrooms'])
logbathrooms = np.log(df_log['bathrooms'])
logsqft_living = np.log(df_log['sqft_living'])
logwaterfront = np.log(df_log['waterfront'])
loggrade = np.log(df_log['grade'])
logyr_renovated = np.log(df_log['yr_renovated'])
logsqft_living15 = np.log(df_log['sqft_living15'])

#Min-Max Scaling
df_log['price'] = (logprice-min(logprice))/(max(logprice)-min(logprice))
df_log['bedrooms'] = (logbedrooms-min(logbedrooms))/(max(logbedrooms)-min(logbedrooms))
df_log['bathrooms'] = (logbathrooms-min(logbathrooms))/(max(logbathrooms)-min(logbathrooms))
df_log['sqft_living'] = (logsqft_living-min(logsqft_living))/(max(logsqft_living)-min(logsqft_living))
df_log['waterfront'] = (logwaterfront-min(logwaterfront))/(max(logwaterfront)-min(logwaterfront))
df_log['grade'] = (loggrade-min(loggrade))/(max(loggrade)-min(loggrade))
df_log['yr_renovated'] = (logyr_renovated-min(logyr_renovated))/(max(logyr_renovated)-min(logyr_renovated))
df_log['sqft_living15'] = (logsqft_living15-min(logsqft_living15))/(max(logsqft_living15)-min(logsqft_living15))

df_log.hist(bins=50, figsize=(20,15), rwidth=.5);
```

```python
# drop waterfront and yr_renovated
df.drop(columns=['waterfront', 'yr_renovated'], axis=1, inplace=True)
df.head()
```
